## 2. Indexing in Pandas


We have the `loc` and `iloc` operation, which are row-first, column-second which is opposite of what we do in native Python.

````python
reviews.iloc[0]
````

### 2.1 iloc

For `iloc`, it stands for **integer-location based indexing**, so it takes arguments in the form of `[row_index, column_index]`.

````python
reviews.iloc[:, 0] # retrieve all row indexes of the first column
````

````python
reviews.iloc[:3, 0]
````

You can even pass a list instead of just `:` operator. The list starts from 0 by the way.

````python
reviews.iloc[[0,1,2], [0,2,5,-3]]
````

````python
reviews.iloc[-5:] # this counts the list 5 elements of the dataframe
````

### 2.2 loc

`.loc` and `.iloc` works same in that it follows this format `[row_index, column_index]` *except* `.loc` uses label-based selection. Meaning, it's the data index value that matters, not its position.

For example, to get the first entry in `reviews`, we would do:

````python
reviews.loc[0, 'country']
````

````python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
````

### 2.3 Difference Between `iloc` and `loc`

Both `iloc` and `loc` are used to select data in a DataFrame, but they follow different indexing rules:

- **`iloc`** uses **Python-style indexing**: the start is **included**, the end is **excluded**.
  - Example: `df.iloc[0:3]` returns rows at index 0, 1, and 2.
  
- **`loc`** uses **label-based indexing**, and both the start and end are **included**.
  - Example: `df.loc[0:3]` returns rows at index 0, 1, 2, and 3.

#### Why This Matters

This can be confusing when the index is numeric. For example:

- `df.iloc[0:1000]` → returns **1000 rows**
- `df.loc[0:1000]` → returns **1001 rows**

`loc` is inclusive because it's also used with labels like strings. For example:

```python
df.loc['Apples':'Potatoes']


````python
reviews.iloc[0:3]
````

````python
reviews.loc[0:3]
````

### 2.4 Conditional Selection

This is where things get interesting for a bit.

Suppose that we're interested specifically in better-than-average wines produced in Italy.

We can startoff by checking if each wine is Italian or not.

````python
reviews.country == "Italy"
````

The result is that it produced a boolean list based on the `country` of each record. The result can be used inside of `.loc` to select the relevant data.

````python
reviews.loc[reviews.country == "Italy"].head()
````

Now, we want to see which of these is better than average. Wine is scored on a scale of 80-100, so we want to see which of these is better-than-average. We can use `&` to combine two conditionals into one.

````python
reviews.loc[(reviews.country == "Italy") & (reviews.points >= 90)].head()
````

Notice that we MUST seperate different conditionals with paranthesis.

We can also use the OR statement.

````python
reviews.loc[(reviews.country == "Italy") | (reviews.points >= 90)].head()
````

### 2.5 Miscellanous

Pandas has some built in conditional selectors to make your life easier. We'll highlight two of them here.

First, it is the `isin()` which lets you to select values whose value is in a list of values. In the following example, we select wine that is only from Italy and France.

````python
reviews.loc[reviews.country.isin(["Italy", "France"])].head()
````

In **pandas**, `.loc` and `.iloc` serve different purposes and accept different types of indexing inputs:

- **`.loc`** is **label-based** indexing. It accepts:
  - Row/column labels
  - Boolean arrays/Series where the index aligns with the axis labels

- **`.iloc`** is **integer position-based** indexing. It accepts:
  - Integer indices or slices
  - **Does NOT accept boolean arrays** because boolean indexing requires label alignment, which `.iloc` does not support

**Reason:**  
Boolean indexing relies on matching the boolean array’s index labels to the DataFrame’s labels. Since `.loc` uses labels, it can align the boolean array properly. In contrast, `.iloc` works purely by integer position without considering labels, so it cannot align a boolean array based on labels and therefore does not support boolean indexing.

---

**Summary:**  
- Use boolean arrays with `.loc` because it aligns labels for filtering.  
- Use integer positions (integers/slices) with `.iloc` because it accesses by positional index only.
